```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000 , stop_words='english' )
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
new = cosine_similarity(vectors)
```

```python
def closest_movies(movie):
    small_df=df[df['title']== movie]
    index=small_df.index[0]
    result = new[index]
    indexed_tuples = list(enumerate(result))
    sorted_tuples = sorted(indexed_tuples, reverse=True , key=lambda x: x[1])
    counter=0
    for i,v in sorted_tuples:
        counter=counter+1
        if counter>6:
            break
        if counter==1:
            continue
        print(df.iloc[i][1])
```

```python
closest_movies('Olympus Has Fallen')
```
